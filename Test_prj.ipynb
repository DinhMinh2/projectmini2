```python
import requests
from bs4 import BeautifulSoup

resp = requests.get('https://webgia.com/tien-ao/')
bfs = BeautifulSoup(resp.content, 'html.parser')

bfs.find('table', class_='coin-main')
rows = bfs.find_all('tr')

for row_id, row in enumerate(rows):
    colums = row.find_all('td')
    if colums:
        if len(colums) >= 7:  # Kiểm tra xem hàng có đủ số ô (cột) cần thiết
            Tien_Ao = colums[1].text.strip()
            Gia = colums[2].get_text(strip=True)  # Corrected this line
            Month = colums[3].text.strip()
            Week = colums[4].text.strip()
            Von_Hoa = colums[5].text.strip()
            Giao_Dich = colums[6].text.strip()

            print(f"TienAo: {Tien_Ao}\
                \nGia: {Gia}\
                \nMonth: {Month}\
                \nWeek: {Week}\
                \nVon_Hoa: {Von_Hoa}\
                \nGiao_Dich: {Giao_Dich}\n\n")
```
